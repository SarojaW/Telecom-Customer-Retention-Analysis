***Exploratory Data Analysis***

```python
#Distribution of churn(yes/no)
sns.countplot(df, x='Churn', color='red')
plt.title("Churn Distribution")
plt.show()
```

*****Significant portion of customers have churned, which suggests areas for improvement. Reducing the churn rate could lead to better customer loyalty and more stable revenue streams.*****

```python
# How many customers fall under each contract type?
sns.countplot(df, x='Contract', color='green')
plt.title("Contract Type Distribution")
plt.show()
```

```python
# What is the distribution of Internet Service types?
sns.countplot(df, x='InternetService', color='blue')
plt.title("Internet Service Type Distribution")
plt.show()
```

```python
# How is the tenure distributed among customers?
sns.histplot(df['tenure'], bins=30, kde=True, color='Orange')
plt.title("Tenure Distribution")
plt.show()
```

*****High churn among new customers(0-5 months). After initial churn customer retention appears more consistent, with more stable count in 20-60 month range. The company may benefit from implementing retention programs targeted at new cutomers to reduce early churn.*****

```python
# What is the distribution of monthly charges?
sns.histplot(df['MonthlyCharges'], bins=30, kde=True, color='Black')
plt.title("Monthly Charges Distribution")
plt.show()
```

*****There is a significant peak at the $20 mark indicating a large number of low-cost plans, but there are customers with higher plans as well. The company could consider segmenting customers by pricing tiers for tailored retention strategies.*****

```python
# What is the distribution of total charges?
sns.histplot(df['TotalCharges'], bins=30, kde=True, color='grey')
plt.title("Distribution of Total Charges")
plt.show()
```

*****Most customers have low total charges. The company couls consider focusing on higher charge tiers for upselling or retention offers to high-value customers.*****

```python
# How does churn vary across different contract types?
sns.countplot(df, x='Contract', hue='Churn', palette='viridis')
plt.title("Churn vs Contract type")
plt.show()
```

*****Focusing on converting month-to-month customers to longer-term contract could reduce churn.*****

```python
# How does churn differ among Internet service types?
sns.countplot(df, x='InternetService', hue='Churn', palette='viridis')
plt.title("Internet Service vs Churn")
plt.show()
```

*****The company could explore customer feedback and issues related to Fiber optic inetrnetservice to reduce churn.*****

```python
# How does churn vary across different payment methods?
plt.figure(figsize=(10,3))
sns.countplot(df, x='PaymentMethod', hue='Churn', palette='rainbow')
plt.title("Payment Method vs Churn")
plt.show()
```

```python
# Are customers with dependents less likely to churn?
sns.countplot(df, x='Dependents', hue='Churn', palette='rainbow')
plt.title("Dependents vs Churn")
plt.show()
```

*****Customers without dependents show higher churn compared to those with dependents.*****

```python
sns.countplot(df, x='TechSupport', hue='Churn', palette='rainbow')
plt.title("TechSupport vs Churn")
plt.show()
```

*****Customers who do not have techh support are more likely to churn.*****

```python
# How do Internet Service, Monthly Charges and Churn interact?
sns.boxplot(df, x='InternetService', y='MonthlyCharges', hue='Churn', palette='viridis')
plt.title("Internet Service vs Monthly Charges vs Churn")
plt.show()
```

*****The box plot shows that customers using fiber optic services have the highest monthly charges and also exhibit higher churn rates. This indicates that pricing or service expectations may be driving churn in premium segments, while lower-cost services like DSL and no internet show relatively lower churn.*****
